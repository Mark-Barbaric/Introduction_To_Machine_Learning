```
import numpy as np
import matplotlib.pyplot as plt

#neural net library
import keras
from keras.models import Sequential 
from keras.layers import Dense

from sklearn.datasets import load_boston
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```
boston = load_boston()
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```
boston_df
```

```
target = boston.target
```

```
target
```

```
#boston_df["target"] = target
```

```
boston_df
```

```
x_train, x_test, y_train, y_test = train_test_split(boston_df, target, train_size=0.7)
```

```
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=20, activation="relu", kernel_initializer='normal', input_shape=x_train.shape[1:]),
    tf.keras.layers.Dense(units=5, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='linear', kernel_initializer='normal')
])
```

```
model.compile(optimizer="adam", loss="mse", metrics=["mse"])

#model.compile(optimizer="rmsprop", loss="mse", metrics=["mse"])
```

```
model.fit(x_train, y_train, epochs=200)
```

```
scores = model.evaluate(x_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
